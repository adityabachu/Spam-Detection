##Remove punctuation

```python
fullCorpus = pd.read_csv("SMSSpamCollection.tsv", sep = '\t')
fullCorpus.columns = ['label', 'body_text']

fullCorpus.head()
```

```python
import string
string.punctuation
```

```python
"I like NLP." == "I like NLP"
```

```python
def remove_punct(text):
  text_nopunct = " ".join([char for char in text if char not in string.punctuation])
  return text_nopunct

fullCorpus['body_text_clean'] = fullCorpus['body_text'].apply(lambda x: remove_punct(x))

fullCorpus.head()
```

##Tokenization

```python
import re

def tokenize(text):
  tokens = re.split('\W+', text)
  return tokens

fullCorpus['body_text_tokenized'] = fullCorpus['body_text_clean'].apply(lambda x: tokenize(x.lower()))

fullCorpus.head()
```

```python
'NLP' == 'nlp'
```

##Remove stopwords

```python
stopword = nltk.corpus.stopwords.words('english')
print(stopword)
```

```python
def remove_stopwords(tokenized_list):
  text = [word for word in tokenized_list if word not in stopword]
  return text

fullCorpus['body_text_nostop'] = fullCorpus['body_text_tokenized'].apply(lambda x: remove_stopwords(x))

fullCorpus.head()
```

```python
ps = nltk.PorterStemmer()
```

```python
dir(ps)
```

##Stemming

```python
print(ps.stem('grows'))
print(ps.stem('growing'))
print(ps.stem('grow'))
```

```python
print(ps.stem('run'))
print(ps.stem('running'))
print(ps.stem('runner'))
```

```python
def clean_text(text):
  text = "".join([word for word in text if word not in string.punctuation])
  tokens = re.split('\W+', text)
  text = [word for word in tokens if word not in stopword]
  return text

fullCorpus['body_text_nostop'] = fullCorpus['body_text'].apply(lambda x: clean_text(x.lower()))

fullCorpus.head()
```

```python
def stemming(tokenized_text):
  text = [ps.stem(word) for word in tokenized_text]
  return text

fullCorpus['body_text_stemmed'] = fullCorpus['body_text_nostop'].apply(lambda x: stemming(x))

fullCorpus.head()
```

##Lemmatization

```python
nltk.download('wordnet')
```

```python
wn = nltk.WordNetLemmatizer()
```

```python
dir(wn)
```

```python
print(ps.stem('meanness'))
print(ps.stem('meaning'))
```

```python
print(wn.lemmatize('meanness'))
print(wn.lemmatize('meaning'))
```

```python
def lemmatizing(tokenized_text):
  text = [wn.lemmatize(word) for word in tokenized_text]
  return text

fullCorpus['body_text_lemmatized'] = fullCorpus['body_text_nostop'].apply(lambda x: lemmatizing(x))

fullCorpus.head()
```

##CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer(analyzer = clean_text)
X_counts = count_vect.fit_transform(fullCorpus['body_text'])
print(X_counts.shape)
print(X_counts)
```

```python
dataset = pd.DataFrame(X_counts.toarray())
dataset
```

##ML Model

###Seperating Dependent and Independent Variables

```python
X = X_counts.toarray()
y = fullCorpus.iloc[:,0].values
```

###Encoding the Dependent Variable

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
```

###Splitting The Dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

###Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```
